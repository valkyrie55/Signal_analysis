# Load a  .wav file and plot its waveform and spectrogram

```python
import soundfile as sf  #for audio conversion
import sounddevice as sd
os.getcwd()
```

```python
audio = 'ravdes\\Actor_01\\03-01-01-01-01-01-01.wav'
data, s_rate = sf.read(audio)
sd.play(data, s_rate)
status = sd.wait()
```

# Waveform

```python
import librosa.display
plt.figure(figsize = (15,5))
librosa.display.waveplot(data, sr = s_rate)
```

# Spectogram

```python
x = librosa.stft(data)
spect = librosa.amplitude_to_db(abs(x))
plt.figure(figsize = (15,5))
librosa.display.specshow(spect, sr = s_rate, x_axis = 'time', y_axis = 'hz')
```
