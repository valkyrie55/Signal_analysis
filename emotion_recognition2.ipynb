```python
x = []
def extract_features(file_path, mfcc, chroma, mel):
    data, s_r = librosa.load(file_path)
    if(chroma):
        stft = np.abs(librosa.stft(data))
    result = np.array([])
    if(mfcc):
        mfccs = np.mean(librosa.feature.mfcc(y = data, sr= s_r, n_mfcc =40).T, axis = 0)
        result = np.hstack((result,mfccs))
    #print(result)
    if(chroma):
        chroma = np.mean(librosa.feature.chroma_stft(S = stft,sr = s_r).T, axis = 0)
        result = np.hstack((result, chroma))
    try:
        if(mel):
            mel = np.mean(librosa.feature.melspectrogram(data,sr = s_r).T, axis = 0)
            result = np.hstack((result, mel))
            x.append(result)
            #print("Done")
    except:
        print("Error")
```

```python
DATADIR = 'C:\\Users\\UPES\\Pictures\\Audio_Speech_analysis\\codes\\pd_data_for_emotion_detection\\ravdes\\'
folders = os.listdir(DATADIR)
for folder in folders:
    path = os.path.join(DATADIR,folder)
    for f in os.listdir(path):
        if f.endswith('.wav'):
            try:
                file_path = os.path.join(path,f)
                    #print(file_path)
                extract_features(file_path,mfcc = 1, chroma = 1, mel = 1)
            except:
                print("Error")
                break
```

```python
from sklearn.neural_network import MLPClassifier
clf1 = MLPClassifier(alpha = 0.01,batch_size = 256,epsilon = 1e-08,
                   hidden_layer_sizes = (300,), learning_rate = 'adaptive', max_iter = 500)
#train model
clf1.fit(x_train, y_train)
#predict on test data
y_pred = clf1.predict(x_test)
# Accuracy score
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_true = y_test,y_pred = y_pred)
print(acc*100)
```
